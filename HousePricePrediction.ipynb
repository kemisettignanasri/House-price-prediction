```python
# Import necessary libraries
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Splitting Data (80% train, 20% test)
X_train, X_test, y_train, y_test = train_test_split(feat, target, test_size=0.2, random_state=42)

# Scaling
scaler = StandardScaler()
X_train_scaler = scaler.fit_transform(X_train)
X_test_scaler = scaler.transform(X_test)

# Decision Tree Model
dt_reg = DecisionTreeRegressor(random_state=42)
dt_reg.fit(X_train_scaler, y_train)
y_pred_dt = dt_reg.predict(X_test_scaler)

# Random Forest Model
rf_reg = RandomForestRegressor(n_estimators=100, random_state=42)
rf_reg.fit(X_train_scaler, y_train)
y_pred_rf = rf_reg.predict(X_test_scaler)

# Evaluation Function
def evaluate_model(name, y_test, y_pred):
    print(f"Model: {name}")
    print("R² Score:", r2_score(y_test, y_pred))
    print("Mean Absolute Error:", mean_absolute_error(y_test, y_pred))
    print("Mean Squared Error:", mean_squared_error(y_test, y_pred))
    print("-" * 50)

# Evaluate Models
evaluate_model("Decision Tree", y_test, y_pred_dt)
evaluate_model("Random Forest", y_test, y_pred_rf)
```

```python
from sklearn.metrics import classification_report
import numpy as np

# Define price categories (Low, Medium, High) based on percentiles
bins = np.percentile(y_train, [33, 66])  # Splitting into 3 equal parts
y_train_class = np.digitize(y_train, bins)  # Convert prices into categories

# Convert predictions into categories using the correct X_train_scaler (created after train_test_split)
y_pred_lin_class = np.digitize(lin_reg.predict(X_train_scaler[:len(y_train)]), bins) # Use X_train_scaler with the same number of samples as y_train
y_pred_sgd_class = np.digitize(sgd_reg.predict(X_train_scaler[:len(y_train)]), bins) # Use X_train_scaler with the same number of samples as y_train
y_pred_dt_class = np.digitize(dt_reg.predict(X_train_scaler[:len(y_train)]), bins)  # Use X_train_scaler with the same number of samples as y_train
y_pred_rf_class = np.digitize(rf_reg.predict(X_train_scaler[:len(y_train)]), bins)  # Use X_train_scaler with the same number of samples as y_train


# Function to print classification report
def print_classification_report(y_true, y_pred, model_name):
    print(f"=== {model_name} Classification Report ===")
    print(classification_report(y_true, y_pred, target_names=["Low", "Medium", "High"]))
    print("=" * 40, "\n")

# Print classification reports for all models
print_classification_report(y_train_class, y_pred_lin_class, "Linear Regression")  # Use y_pred_lin_class (Categorized predictions)
print_classification_report(y_train_class, y_pred_sgd_class, "SGD Regressor")   # Use y_pred_sgd_class (Categorized predictions)
print_classification_report(y_train_class, y_pred_dt_class, "Decision Tree")    # Use y_pred_dt_class (Categorized predictions)
print_classification_report(y_train_class, y_pred_rf_class, "Random Forest")    # Use y_pred_rf_class (Categorized predictions)
```
